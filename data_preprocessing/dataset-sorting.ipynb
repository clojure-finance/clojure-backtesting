```python
import os
import pandas as pd

filename = os.path.join('CRSP-extract.csv')
```

```python
df = pd.read_csv(filename, dtype={'CFACPR': 'float64',
       'CFACSHR': 'float64',
       'CUSIP': 'object',
       'HSICCD': 'object',
       'NCUSIP': 'object'})
df.head()
```

```python
df = df.sort_values(['date', 'TICKER'], ascending=[True, True])
df.head()
```

```python
df.to_csv('CRSP-extract-sorted.csv', index=False)
```
